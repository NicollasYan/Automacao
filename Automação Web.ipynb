Nesse código eu automatizei o chrome para pequisar as cotaçôes do Dólar, Euro e do Ouro para atualizar os valores dos produtos importados de uma planilha.

Para automatizar esse código eu utilizei o webdriver e o Selenium (que são ferramentas open source para automatização web)

E utilizei o jupyter para desenvolver o código

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

# Abrindo Chrome
nav = webdriver.Chrome()

# Pesquisando a cotação do dolar
nav.get("https://www.google.com.br/")
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('Cotação Dólar')
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)
cotacao_dolar = nav.find_element_by_xpath('//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_dolar)

# Pesquisando a cotação do euro
nav.get("https://www.google.com.br/")
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('Cotação Euro')
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)
cotacao_euro = nav.find_element_by_xpath('//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_euro)

#Pesquisando a cotação do ouro
nav.get("https://www.melhorcambio.com/ouro-hoje#:~:text=O%20valor%20do%20grama%20do,em%20R%24%20292%2C00.")
cotacao_ouro = nav.find_element_by_xpath('//*[@id="comercial"]').get_attribute('value')

# Aqui eu tive que mudar a vírgula do valor para um ponto pro python reconhecer como valor
cotacao_ouro = cotacao_ouro.replace(",", ".")
print(cotacao_ouro)

nav.quit()
```

Agora eu irei importar uma base de dados básica e utilizarei a blibioteca pandas para trabalhar com ela

```python
import pandas as pd
# Lendo o arquivo
tabela = pd.read_excel("Produtos.xlsx")
# Eu poderia usar o (print) para exibir a planilha mas a visualização ficaria ruim
display(tabela)
```

Agora é só atualizar os valores 

```python
tabela.loc[tabela["Moeda"] == "Dólar", "Cotação"] = float(cotacao_dolar)
tabela.loc[tabela["Moeda"] == "Euro", "Cotação"] = float(cotacao_euro)
tabela.loc[tabela["Moeda"] == "Ouro", "Cotação"] = float(cotacao_ouro)
tabela["Preço Base Reais"] = tabela["Cotação"] * tabela["Preço Base Original"]
tabela["Preço Final"] = tabela["Preço Base Reais"] * tabela["Ajuste"]
display(tabela)
```

Agora só exportar a tabela atualizada

```python
tabela.to_excel("Produtos Atualizados.xlsx")
```
